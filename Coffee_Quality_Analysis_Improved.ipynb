# 7. Data Cleaning

## i. Remove Unnecessary Columns

 The dataset has technical/duplicate columns (e.g., Unnamed: 0, duplicate Owner column).

```python
coffee_df.drop(columns=[
    "Unnamed: 0", "unit_of_measurement",
    "Certification.Body", "Certification.Address",
    "Certification.Contact", "Expiration"
], inplace=True)
```

## ii. Handle Missing Values

Categorical variables â†’ fill with mode (most common).

Numerical variables â†’ fill with mean/median.

```python
# Drop high-missing columns
coffee_df.drop(columns=["Lot.Number","Farm.Name","Company","Producer"], inplace=True)

# Fill numeric NaNs with median
num_cols = coffee_df.select_dtypes(include=['float64','int64']).columns
coffee_df[num_cols] = coffee_df[num_cols].fillna(coffee_df[num_cols].median())

# Fill categorical NaNs with mode
cat_cols = coffee_df.select_dtypes(include=['object']).columns
for col in cat_cols:
    coffee_df[col].fillna(coffee_df[col].mode()[0], inplace=True)
```

## iii. Handle Inconsistent Formatting

Example: Altitude column has ranges and text â†’ weâ€™ll keep numeric altitude columns instead.

```python
# Drop the inconsistent Altitude text column
coffee_df.drop(columns=['Altitude'], inplace=True, errors='ignore')

# Ensure altitude columns are numeric
altitude_cols = ['altitude_low_meters', 'altitude_high_meters', 'altitude_mean_meters']
coffee_df[altitude_cols] = coffee_df[altitude_cols].apply(pd.to_numeric, errors='coerce')
```

## iv. Fix Outliers

Some values (e.g., Aroma = 200, Cup Points = 0) are unrealistic â†’ remove or cap.

```python
# Define quality variables
quality_cols = ["Aroma", "Flavor", "Aftertaste", "Acidity", "Body", 
                "Balance", "Uniformity", "Clean.Cup", "Sweetness", "Total.Cup.Points"]

# Remove unrealistic values (e.g., scores must be between 0 and 100)
for col in quality_cols:
    coffee_df = coffee_df[(coffee_df[col] >= 0) & (coffee_df[col] <= 100)]
```

## v. Standardize Text Columns

Ensure categorical variables have consistent formatting.

```python
# Example: Standardize country names and species
coffee_df['Country.of.Origin'] = coffee_df['Country.of.Origin'].str.strip().str.title()
coffee_df['Species'] = coffee_df['Species'].str.strip().str.capitalize()
```

## vi. Remove Duplicates

```python
coffee_df.drop_duplicates(inplace=True)
```

## vii. Final Check

```python
print("Final Dataset Shape:", coffee_df.shape)
print("Remaining Missing Values:\n", coffee_df.isnull().sum())
```

```python
# ðŸ“Œ DESCRIPTIVE STATISTICS (VARIABLE-WISE)

# 1ï¸âƒ£ Aroma
print("Aroma Statistics:")
print("Mean:", coffee_df['Aroma'].mean())
print("Median:", coffee_df['Aroma'].median())
print("Standard Deviation:", coffee_df['Aroma'].std())
print("Minimum:", coffee_df['Aroma'].min())
print("Maximum:", coffee_df['Aroma'].max())
print("-"*40)

# 2ï¸âƒ£ Flavor
print("Flavor Statistics:")
print("Mean:", coffee_df['Flavor'].mean())
print("Median:", coffee_df['Flavor'].median())
print("Standard Deviation:", coffee_df['Flavor'].std())
print("Minimum:", coffee_df['Flavor'].min())
print("Maximum:", coffee_df['Flavor'].max())
print("-"*40)

# 3ï¸âƒ£ Acidity
print("Acidity Statistics:")
print("Mean:", coffee_df['Acidity'].mean())
print("Median:", coffee_df['Acidity'].median())
print("Standard Deviation:", coffee_df['Acidity'].std())
print("Minimum:", coffee_df['Acidity'].min())
print("Maximum:", coffee_df['Acidity'].max())
print("-"*40)

# 4ï¸âƒ£ Total Cup Points
print("Total Cup Points Statistics:")
print("Mean:", coffee_df['Total.Cup.Points'].mean())
print("Median:", coffee_df['Total.Cup.Points'].median())
print("Standard Deviation:", coffee_df['Total.Cup.Points'].std())
print("Minimum:", coffee_df['Total.Cup.Points'].min())
print("Maximum:", coffee_df['Total.Cup.Points'].max())
print("-"*40)

# 5ï¸âƒ£ Moisture
print("Moisture Statistics:")
print("Mean:", coffee_df['Moisture'].mean())
print("Median:", coffee_df['Moisture'].median())
print("Standard Deviation:", coffee_df['Moisture'].std())
print("Minimum:", coffee_df['Moisture'].min())
print("Maximum:", coffee_df['Moisture'].max())
print("-"*40)
```

## Hypothesis Testing

```python
# ðŸ“Œ Hypothesis Testing : T Test
arabica = coffee_df[coffee_df['Species'] == 'Arabica']['Total.Cup.Points'].dropna()
robusta = coffee_df[coffee_df['Species'] == 'Robusta']['Total.Cup.Points'].dropna()

t_stat, p_val = stats.ttest_ind(arabica, robusta, equal_var=False)

print("\nHypothesis Test: Arabica vs Robusta Cup Scores")
print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("âœ… Significant difference found between Arabica and Robusta cup scores.")
else:
    print("âŒ No significant difference found between Arabica and Robusta cup scores.")

```

## 10. Insights and Recommendations

1. Enhance Key Quality Attributes

Finding: Aroma, Flavor, and Acidity show the strongest positive correlations with Total Cup Points.

Recommendation: Producers should focus on post-harvest processing techniques such as controlled fermentation, improved drying methods, and selective picking of ripe cherries to enhance these attributes.

2. Maintain Optimal Moisture Content

Finding: Average moisture is within the ideal range (~10%), but extremes (below 8% or above 12%) can harm quality and shelf life.

Recommendation: Standardize drying practices and adopt climate-controlled storage facilities to ensure beans remain within the safe moisture range.

3. Leverage Altitude for Quality Improvement

Finding: Higher-altitude farms (>1500m) generally achieve better cup scores.

Recommendation: Lower-altitude farms can adapt shade-grown methods, improved soil fertility, and careful harvesting practices to partially replicate the benefits of high-altitude conditions.

4. Species Diversification and Innovation

Finding: Arabica outperforms Robusta significantly in cup scores.

Recommendation:

For Arabica farms: Continue focusing on specialty-grade quality for premium markets.

For Robusta farms: Explore blends with Arabica or disease-resistant Arabica hybrids to improve market value without sacrificing resilience.
